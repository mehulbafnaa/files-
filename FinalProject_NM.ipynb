```python
import numpy as np
import pandas as  pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score


url = 'https://raw.githubusercontent.com/mehulbafnaa/files1/main/allwine.csv'
df = pd.read_csv(url)
```

```python


class LogisticRegression:
    def __init__(self):
        self.w = None
        self.b = None
        self.costs = None

    def weightInitialization(self, n_features):
        self.w = np.zeros((1, n_features))
        self.b = 0

    def sigmoid_activation(self, result):
        final_result = 1 / (1 + np.exp(-result))
        return final_result

    def model_optimize(self, X, Y):
        m = X.shape[0]

        # Prediction
        final_result = self.sigmoid_activation(np.dot(self.w, X.T) + self.b)
        Y_T = Y.T
        cost = (-1 / m) * (np.sum((Y_T * np.log(final_result)) + ((1 - Y_T) * (np.log(1 - final_result)))))

        # Gradient calculation
        dw = (1 / m) * (np.dot(X.T, (final_result - Y.T).T))
        db = (1 / m) * (np.sum(final_result - Y.T))

        grads = {"dw": dw, "db": db}

        return grads, cost

    def model_predict(self, X, Y, learning_rate, no_iterations):
        self.costs = []

        for i in range(no_iterations):
            grads, cost = self.model_optimize(X, Y)
            dw = grads["dw"]
            db = grads["db"]

            # Weight update
            self.w = self.w - (learning_rate * dw.T)
            self.b = self.b - (learning_rate * db)

            if (i % 100 == 0):
                self.costs.append(cost)

        coeff = {"w": self.w, "b": self.b}
        gradient = {"dw": dw, "db": db}

        return self.costs

    def predict(self, X):
        final_result = self.sigmoid_activation(np.dot(self.w, X.T) + self.b)
        y_pred = np.where(final_result > 0.5, 1, 0)
        return y_pred
```

```python
X = df[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
          'chlorides', 'free sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']].values
y = df['quality'].values

# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

X_tr_arr = X_train
X_ts_arr = X_test
y_tr_arr = y_train
y_ts_arr = y_test


print('Input Shape', (X_tr_arr.shape))
print('Output Shape', X_test.shape)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score

# Assuming the LogisticRegression class has been defined as per your previous message

# Create an instance of the LogisticRegression
model = LogisticRegression()

# Get number of features
n_features = X_tr_arr.shape[1]
print('Number of Features', n_features)

# Initialize weights and biases within the model
model.weightInitialization(n_features)

# Train the model using gradient descent
costs = model.model_predict(X_tr_arr, y_tr_arr, learning_rate=0.0001, no_iterations=4500)

# Get the optimized weights and bias from the model
w = model.w
b = model.b
print('Optimized weights', w)
print('Optimized intercept', b)

# Predict on the training and test set
y_tr_pred = model.predict(X_tr_arr)
y_ts_pred = model.predict(X_ts_arr)

# Calculate accuracy
training_accuracy = accuracy_score(y_tr_pred.flatten(), y_tr_arr)
test_accuracy = accuracy_score(y_ts_pred.flatten(), y_ts_arr)

print('Training Accuracy:', training_accuracy)
print('Test Accuracy:', test_accuracy)
```

```python

import numpy as np

class LogisticRegression1:
    def __init__(self):
        self.w = None
        self.b = None
        self.costs = []

    def weightInitialization(self, n_features):
        self.w = np.zeros((1, n_features))
        self.b = 0

    def sigmoid_activation(self, result):
        return 1 / (1 + np.exp(-result))

    def model_optimize(self, X, Y):
        m = X.shape[0]
        final_result = self.sigmoid_activation(np.dot(self.w, X.T) + self.b)
        cost = (-1 / m) * np.sum((Y.T * np.log(final_result)) + ((1 - Y.T) * np.log(1 - final_result)))
        dw = (1 / m) * np.dot(X.T, (final_result - Y.T).T)
        db = (1 / m) * np.sum(final_result - Y.T)
        return {"dw": dw, "db": db}, cost

    def train(self, X, Y, learning_rate, no_iterations):
        self.weightInitialization(X.shape[1])
        for i in range(no_iterations):
            grads, cost = self.model_optimize(X, Y)
            self.w -= learning_rate * grads['dw'].T
            self.b -= learning_rate * grads['db']
            if i % 100 == 0:
                self.costs.append(cost)

    def predict(self, X):
        final_result = self.sigmoid_activation(np.dot(self.w, X.T) + self.b)
        return np.where(final_result > 0.5, 1, 0)

class DynamicEnsembleLogisticModel:
    def __init__(self):
        self.LR_middle = LogisticRegression1()
        self.LR_left = LogisticRegression1()
        self.LR_right = LogisticRegression1()

    def train(self, X, Y, learning_rate=0.01, no_iterations=1000):
        # This might be modified based on the specific data and requirements
        self.LR_middle.train(X, Y, learning_rate, no_iterations)
        self.LR_left.train(X, Y, learning_rate, no_iterations)
        self.LR_right.train(X, Y, learning_rate, no_iterations)

    def predict(self, X):
        prob_m = self.LR_middle.sigmoid_activation(np.dot(self.LR_middle.w, X.T) + self.LR_middle.b)
        prob_l = self.LR_left.sigmoid_activation(np.dot(self.LR_left.w, X.T) + self.LR_left.b)
        prob_r = self.LR_right.sigmoid_activation(np.dot(self.LR_right.w, X.T) + self.LR_right.b)

        prob_1 = prob_l * prob_m + prob_r * (1 - prob_m)
        prob_0 = (1 - prob_l) * prob_m + (1 - prob_r) * (1 - prob_m)

        return np.where(prob_1 > prob_0, 1, 0).flatten()

# Example usage:
# X, Y = load_your_data_somehow()
# model = DynamicEnsembleLogisticModel()
# model.train(X, Y)
# predictions = model.predict(X)


# Example Usage
model1 = DynamicEnsembleLogisticModel()
model1.train(X_tr_arr, y_tr_arr)  # Train the model on your training data
predictions = model1.predict(X_ts_arr)  # Predict on new data

# You would typically use accuracy or another metric to evaluate the model
from sklearn.metrics import accuracy_score
print("Accuracy:", accuracy_score(y_ts_arr, predictions))
```

```python


class LogisticRegressionNode:
    def __init__(self):
        self.weights = None
        self.bias = None

    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    def train(self, X, y, learning_rate, epochs, reg_lambda=0.01):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        for _ in range(epochs):
            linear_model = np.dot(X, self.weights) + self.bias
            y_predicted = self.sigmoid(linear_model)
            dw = (1 / n_samples) * np.dot(X.T, (y_predicted - y)) + reg_lambda * self.weights
            db = (1 / n_samples) * np.sum(y_predicted - y)
            self.weights -= learning_rate * dw
            self.bias -= learning_rate * db

    def predict_proba(self, X):
        linear_model = np.dot(X, self.weights) + self.bias
        return self.sigmoid(linear_model)



class DeepDynamicEnsembleModel:
    def __init__(self):
        self.nodes = {i: LogisticRegressionNode() for i in range(1, 8)}

    def fit(self, X, y, learning_rate=0.01, epochs=100):
        # Train root node
        self.nodes[1].train(X, y, learning_rate, epochs)
        prob_1 = self.nodes[1].predict_proba(X)

        # Second layer training using probabilities from the root node
        self.nodes[2].train(X, y * prob_1, learning_rate, epochs)
        self.nodes[3].train(X, y * (1 - prob_1), learning_rate, epochs)

        # Calculate probabilities for nodes in the second layer
        prob_2 = self.nodes[2].predict_proba(X)
        prob_3 = self.nodes[3].predict_proba(X)

        # Third layer training using probabilities from the second layer
        self.nodes[4].train(X, y * prob_2, learning_rate, epochs)
        self.nodes[5].train(X, y * (1 - prob_2), learning_rate, epochs)
        self.nodes[6].train(X, y * prob_3, learning_rate, epochs)
        self.nodes[7].train(X, y * (1 - prob_3), learning_rate, epochs)

    def predict(self, X):
        # Aggregate probabilities across the model's structure
        prob_1 = self.nodes[1].predict_proba(X)
        prob_2 = self.nodes[2].predict_proba(X) * prob_1
        prob_3 = self.nodes[3].predict_proba(X) * (1 - prob_1)
        prob_4 = self.nodes[4].predict_proba(X) * prob_2
        prob_5 = self.nodes[5].predict_proba(X) * (1 - prob_2)
        prob_6 = self.nodes[6].predict_proba(X) * prob_3
        prob_7 = self.nodes[7].predict_proba(X) * (1 - prob_3)

        # Weighted sum from all leaves
        final_prob = prob_4 + prob_5 + prob_6 + prob_7
        return np.where(final_prob > 0.5, 1, 0)  # Convert probabilities to class labels



model2 = DeepDynamicEnsembleModel()

model2.fit(X_tr_arr, y_tr_arr)

predictions1 = model1.predict(X_ts_arr)  # Predict on new data

print("Accuracy:", accuracy_score(y_ts_arr, predictions1))



```
